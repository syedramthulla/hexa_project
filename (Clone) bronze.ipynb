```python
from pyspark.sql.functions import *
from pyspark.sql.functions import types

df=spark.read.csv("path_to_csv").option("header","true").option("inferSchema","true")
display(df)

df=spark.read.format("parquet").load("path_to_parquet")
display(df)

df=spark.read.json("path_to_json")
display(df)

df=spark.read.format("delta").load("path_to_delta")
display(df)

df=spark.write.format("delta").mode("overwrite").save("path_to_delta")
display(df)


df=spark.write.format("delta").mode("appened").save("path_to_delta")
display(df)
 
df1=df.dropDuplicates()
display(df1)
```
