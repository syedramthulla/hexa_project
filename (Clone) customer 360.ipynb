## Step 1: Create DataFrame with schema

```python
from pyspark.sql.types import *
from pyspark.sql.functions import current_timestamp
from datetime import date

schema = StructType([
    StructField("customer_id", StringType(), True),
    StructField("first_name", StringType(), True),
    StructField("last_name", StringType(), True),
    StructField("email", StringType(), True),
    StructField("phone_number", StringType(), True),
    StructField("city", StringType(), True),
    StructField("state", StringType(), True),
    StructField("country", StringType(), True),

    # Business fields
    StructField("customer_status", StringType(), True),
    StructField("loyalty_tier", StringType(), True),
    StructField("registration_date", StringType(), True),

    # Personalization
    StructField("preferred_category", StringType(), True),
    StructField("preferred_channel", StringType(), True),

    # SCD Type 2
    StructField("start_date", DateType(), True),
    StructField("end_date", DateType(), True),
    StructField("is_current", BooleanType(), True),

    # Audit
    StructField("created_at", TimestampType(), True),
    StructField("updated_at", TimestampType(), True)
])
```

## Step 2: Create empty DataFrame

```python
df_sample = spark.createDataFrame(data, schema)
```

## Step 3: Write as Delta Table

### # Option 1: Save as path

```python
df_sample.write.format("delta") \
    .mode("overwrite") \
    .save("/Volumes/deltalakesyed/default/volumes/customers")
```
